```python
#--------------------------------------------------------------------------------------
# fun: get playerinfo by interid in playerinfo2
# input: inter id
# outout: {'interid','playerrole','playerid'}
#--------------------------------------------------------------------------------------
def getplayerinfo_byinterid(interid):
    db=mysql.connector.connect(host="localhost",
                           database='lccgame',
                           user='host',
                           password='host')
    cursor=db.cursor()
    cursor.execute("SELECT userid FROM playerinfo2 WHERE interid=%r" %interid)
    playerid=cursor.fetchall()
    cursor.execute("SELECT playerrole FROM playerinfo2 WHERE interid=%r" %interid)
    playerrole=cursor.fetchall()
    db.close()
    playerinfo_list=[]
    for i in range(len(playerid)):
        playerinfo_list.append({
            'interid':interid,
            'playerid':playerid[i][0],
            'playerrole':playerrole[i][0]
        })
    return playerinfo_list


#--------------------------------------------------------------------------------------
# fun: get gamemsgs by interid in gamemsgs2
# input: interid
# outout: {'interid','playerrole','playerid'}
#--------------------------------------------------------------------------------------
def getgamemsgs_byinterid(interid):
    db=mysql.connector.connect(host="localhost",
                           database='lccgame',
                           user='host',
                           password='host')
    cursor=db.cursor()
    cursor.execute("""SELECT gameid,protocolid,msgsenderid,msgsenderrole,msgreceiverid,msgreceiverrole,msgbody
                    FROM gamemsgs2 
                    WHERE interid=%r""" %interid)
    gamemsgsdata=cursor.fetchall()
    db.close()
    gamemsgs_list=[]
    for i in range(len(gamemsgsdata)):
        gamemsgs_list.append({
            'interid':interid,
            'gameid':gamemsgsdata[i][0],
            'protocol':gamemsgsdata[i][1],
            'msid':gamemsgsdata[i][2],
            'msrole':gamemsgsdata[i][3],
            'mrid':gamemsgsdata[i][4],
            'mrrole':gamemsgsdata[i][5],
            'mbody':gamemsgsdata[i][6]
        })
    return gamemsgs_list
```
